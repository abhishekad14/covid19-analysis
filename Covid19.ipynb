```python
#import modules
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 

print('Modules are imported.')
```

```python
#import dataset
cov_df_csv= pd.read_csv('covid19_Confirmed_dataset.csv')
cov_df_csv.head(10)
```

```python
cov_df_csv.shape
```

```python
#drop irrelevant columns
cov_df_csv.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
cov_df_csv.head(10)
```

Aggregating country and provinces 

```python
cov_df_agg=cov_df_csv.groupby("Country/Region").sum()
```

```python
cov_df_agg.head()
```

```python
cov_df_agg.shape
```

```python
cov_df_agg.loc["India"].diff().plot()
```

<b> Calculating spread rate 

```python
print("India Maximum infection=",cov_df_agg.loc["India"].diff().max())
print("China Maximum infection=",cov_df_agg.loc["China"].diff().max())
print("Italy Maximum infection=",cov_df_agg.loc["Italy"].diff().max())
print("France Maximum infection=",cov_df_agg.loc["France"].diff().max())
print("Spain Maximum infection=",cov_df_agg.loc["Spain"].diff().max())
```

<b> Maximum virus spread rate worldwide

```python
countries = list(cov_df_agg.index)
max_spread_rate = []
for country in countries :
    max_spread_rate.append(cov_df_agg.loc[country].diff().max())
cov_df_agg['max spread rate'] = max_spread_rate
```

```python
cov_df_agg.head(8)
```

<b>Dataframe with max virus spread rate

```python
cov_df_agg1 = pd.DataFrame(cov_df_agg['max spread rate'])
```

```python
cov_df_agg1.head()
```
